```python
#We need to clean up the data and remove any unnecessary things such as removing rows that don't contain actual
#country. The World Bank provides meta_data on whichrows are countries and which are aggregates. So we first 
#have to import another csv file name MetaData_Country
meta_data = pd.read_csv("MetaData_Country.csv", encoding = "utf-8")
```

```python
#show the first ten rows of the meta data
meta_data.head(10)
```

```python
#merge the first csv file data_set with meta_data on the key 'Country Code'
merge = pd.merge(data_set, meta_data, on = "Country Code")
merge.shape
```

```python
#Show the first ten rows of the new merge file and you can see that the two data set merged with each other
merge.head(10)
```

```python
#The region values show 'NaN' when the row is not an actual country. So now we must remove rows where region is NaN
merge = merge[pd.notnull(merge['Region'])]
```

```python
merge.head(2)
```

```python
#we can also remove columns with irrelevant information such as the columns "Indicator Name" and "Indicator Code"
merge.columns
```

```python
merge = merge.drop(merge.columns[[60,61,64,65]], axis=1)
merge = merge.drop('Indicator Name', 1)
merge = merge.drop('Indicator Code', 1)
```

```python
#show the new data set without the two irrelevant columns
merge.columns
```

```python
#Count the number of rows for each year. Note that 'NaN' values will not get counted towards the total
merge.count()
```

```python
#Check years that seems to have rows containing no data
merge['2015']
```

```python
merge['2018']
```

```python
merge['2019']
```

```python
#remove the colums that has no row data
merge = merge.drop(['2015','2018','2019'], 1)
```

```python
#check that those columns have been removed
merge.count()
```

```python
#Now we should see which countries have no data for any year. We can use row means to find which countries have
#no data. We can take the means for each row on axis 1. 
merge.mean(axis=1) #this code takes the mean of all the numeric quantities by row
```

```python
#we can see that NaN appears when there is no data for that row. Now we can remove rows where there are no values 
#in any year
merge = merge[pd.notnull(merge.mean(axis=1))]
```

```python
#show the first 10 rows of the new data
merge.head(10)
```

```python
my_new_df = merge
```
