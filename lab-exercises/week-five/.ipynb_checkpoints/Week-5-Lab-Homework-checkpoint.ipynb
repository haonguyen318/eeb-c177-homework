#### 4. Knowing that two strings can be concatenated using the + operator, write a loop that takes a string and produces a new string with the characters in reverse order, so ‘Ucla’ becomes ‘alcU’.

```python
characters = "ucla"
reverse_order = "" #create an empty string to store the characters in reverse order
for n in characters:
    reverse_order = n + reverse_order
print(reverse_order)
```

#### 5. Write a loop using enumerate(coefs) which computes the value y of any polynomial, given x and coefs.

```python
x = 5
coefs = [2, 4, 3]
y = coefs[0] * x**0 + coefs[1] * x**1 + coefs[2] * x**2
print(y)
```

```python
x = 5
coefs = [2, 4, 3]
for idx, val in enumerate(coefs):
    y = coefs[0] * x**0 + coefs[1] * x**1 + coefs[2] * x**2
print(y)
```

```python
x = 5
coefs = [2, 4, 3]
y = 0 
for idx, val in enumerate(coefs):
    y += val * x**idx
print(y)
```

### Part 3 (Lists)

#### 1. Explain what a list is.

A list is an ordered collection of values of the same or different
types where each value can appear multiple times. 

#### 2. Explain how to create and index lists of simple values.

The list function works by separating each value with commas and 
putting these values in a pair of square brackets. In addition,
you can access the elements of a list using the index of the element.

#### 3. Explain how to change the values of individual elements.

#### 2. Write a loop that counts the number of vowels in a character string. Test it on a few individual words and full sentences.

```python
vowels = ['a','e','i','o','u']

example1 = "The cat in the hat."
count = 0
for counts in example1:
    if counts in vowels:
        count += 1
print(count)
```

```python
example2 = "I like green eggs and ham."
count = 0
for counts in example2:
    if counts in vowels:
        count += 1
print(count)
```
